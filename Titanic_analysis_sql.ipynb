```python
# !pip install pandasql
from pandasql import sqldf
sql = lambda q : sqldf(q, globals())
```

```python
sql(" SELECT * FROM file limit 2 ") # select statement
```

```python
sql(" SELECT count(*) FROM file ") # number of records
```

```python
sql(" SELECT distinct sex FROM file ") # distinct gives unique category counts in the column
```

```python
sql("SELECT distinct pclass FROM file")
```

```python
sql(" SELECT sex, count(*) FROM file group by sex ")
```

```python
sql("SELECT pclass,count(*) FROM file group by pclass ")
```

```python
sql(" SELECT pclass, count(*) FROM file group by pclass ")
```

```python
sql(" SELECT pclass, sex, count(*) FROM file group by pclass, sex ")
```
